SMART GRID LOAD DECISION AGENT

```python
# ============================================================
# SMART GRID LOAD DECISION AGENT
# AI BASED CFAI PROJECT
# CO1 + CO2 + CO3
# ============================================================

from dataclasses import dataclass, field
from typing import Dict, List, Tuple

# ============================================================
# CO1 - AI PROBLEM REPRESENTATION
# ============================================================

@dataclass
class Consumer:
    consumer_id: str
    area: str

@dataclass
class GridNode:
    node_id: str
    capacity: int

# Consumer Data

CONSUMERS = [
    Consumer("C101", "North"),
    Consumer("C102", "South"),
    Consumer("C103", "East"),
    Consumer("C104", "North")
]

# Grid Node Data

GRID_NODES = {
    "G1": GridNode("G1", 2),
    "G2": GridNode("G2", 2)
}

# ============================================================
# STATE REPRESENTATION
# ============================================================

@dataclass
class State:

    allocations: Dict[str, List[Tuple[str, str]]] = field(
        default_factory=dict
    )

# ============================================================
# SMART GRID PROBLEM
# ============================================================

class SmartGridProblem:

    def __init__(self, consumers, nodes):

        self.consumers = consumers
        self.nodes = nodes

    # ========================================================
    # CO3 - CONSTRAINT CHECKING
    # ========================================================

    def is_safe(self, node, area, state):

        # Constraint 1:
        # Same area consumers should not use same node

        for consumer in state.allocations[node]:

            if consumer[1] == area:

                return False

        # Constraint 2:
        # Node capacity should not exceed

        if len(state.allocations[node]) >= self.nodes[node].capacity:

            return False

        return True

    # ========================================================
    # GOAL TEST
    # ========================================================

    def is_goal(self, index):

        return index == len(self.consumers)

# ============================================================
# CO2 - DFS SEARCH ALGORITHM
# ============================================================

def dfs(problem, state, index):

    if problem.is_goal(index):

        print("\nSUCCESS: LOAD DISTRIBUTION COMPLETED")
        return True

    consumer = problem.consumers[index]

    print("\n------------------------------------------------")
    print("Processing Consumer:", consumer.consumer_id)
    print("Area:", consumer.area)
    print("------------------------------------------------")

    for node in problem.nodes:

        print("\nChecking Grid Node:", node)

        if problem.is_safe(node, consumer.area, state):

            print("Safe Allocation Possible")

            state.allocations[node].append(
                (consumer.consumer_id, consumer.area)
            )

            print(
                "Allocated",
                consumer.consumer_id,
                "to",
                node
            )

            # Recursive DFS Search

            if dfs(problem, state, index + 1):

                return True

            # =================================================
            # CO3 - BACKTRACKING
            # =================================================

            removed_consumer = state.allocations[node].pop()

            print("Backtracking Performed")

            print(
                "Removed Consumer:",
                removed_consumer[0]
            )

        else:

            print("Constraint Violated")
            print("Cannot Allocate Consumer")

    return False

# ============================================================
# DISPLAY FUNCTIONS
# ============================================================

def show_consumers():

    print("\n================================================")
    print("CONSUMER DETAILS")
    print("================================================")

    for consumer in CONSUMERS:

        print(
            "Consumer ID:",
            consumer.consumer_id,
            "| Area:",
            consumer.area
        )

def show_nodes():

    print("\n================================================")
    print("GRID NODE DETAILS")
    print("================================================")

    for node_id, node in GRID_NODES.items():

        print(
            "Node:",
            node_id,
            "| Capacity:",
            node.capacity
        )

# ============================================================
# FINAL OUTPUT
# ============================================================

def show_final_distribution(state):

    print("\n================================================")
    print("FINAL LOAD DISTRIBUTION")
    print("================================================")

    for node in state.allocations:

        print("\nGrid Node:", node)

        if state.allocations[node]:

            for consumer in state.allocations[node]:

                print(
                    "Consumer:",
                    consumer[0],
                    "| Area:",
                    consumer[1]
                )

        else:

            print("No Consumers Allocated")

# ============================================================
# MAIN PROGRAM
# ============================================================

def main():

    print("\n================================================")
    print("SMART GRID LOAD DECISION AGENT")
    print("AI BASED CFAI PROJECT")
    print("CO1 + CO2 + CO3")
    print("================================================")

    # Display Input Data

    show_consumers()
    show_nodes()

    # Initial State

    initial_state = State(
        allocations={
            "G1": [],
            "G2": []
        }
    )

    # Create Problem

    problem = SmartGridProblem(
        CONSUMERS,
        GRID_NODES
    )

    print("\n================================================")
    print("DFS SEARCH + CSP STARTED")
    print("================================================")

    # Start DFS Search

    success = dfs(
        problem,
        initial_state,
        0
    )

    # Final Output

    if success:

        print("\nLOAD DISTRIBUTION GENERATED SUCCESSFULLY")

        show_final_distribution(initial_state)

    else:

        print("\nFAILED TO GENERATE LOAD DISTRIBUTION")

    print("\n================================================")
    print("PROJECT COMPLETED SUCCESSFULLY")
    print("================================================")

# ============================================================
# DRIVER CODE
# ============================================================

if __name__ == "__main__":

    main()
```
