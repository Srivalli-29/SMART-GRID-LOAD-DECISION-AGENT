```python
# ============================================================
# SMART GRID LOAD DECISION AGENT
# AI BASED CFAI PROJECT
# CO1 + CO2 + CO3 + CO4 + CO5 + CO6
# ============================================================

from dataclasses import dataclass, field
from typing import Dict, List, Tuple

# ============================================================
# CO1 - AI PROBLEM REPRESENTATION
# ============================================================

@dataclass
class Consumer:
    consumer_id: str
    area: str

@dataclass
class GridNode:
    node_id: str
    capacity: int

# Consumer Data

CONSUMERS = [
    Consumer("C101", "North"),
    Consumer("C102", "South"),
    Consumer("C103", "East"),
    Consumer("C104", "North"),
    Consumer("C105", "West"),
    Consumer("C106", "South")
]

# Grid Node Data

GRID_NODES = {
    "G1": GridNode("G1", 2),
    "G2": GridNode("G2", 2),
    "G3": GridNode("G3", 2)
}

# ============================================================
# STATE REPRESENTATION
# ============================================================

@dataclass
class State:

    allocations: Dict[str, List[Tuple[str, str]]] = field(
        default_factory=dict
    )

# ============================================================
# SMART GRID PROBLEM
# ============================================================

class SmartGridProblem:

    def __init__(self, consumers, nodes):

        self.consumers = consumers
        self.nodes = nodes

    # ========================================================
    # CO3 - CONSTRAINT CHECKING
    # ========================================================

    def is_safe(self, node, area, state):

        # Constraint 1:
        # Same area consumers should not overload same node

        for consumer in state.allocations[node]:

            if consumer[1] == area:

                return False

        # Constraint 2:
        # Node capacity should not exceed

        if len(state.allocations[node]) >= self.nodes[node].capacity:

            return False

        return True

    # ========================================================
    # GOAL TEST
    # ========================================================

    def is_goal(self, index):

        return index == len(self.consumers)

# ============================================================
# CO2 - DFS SEARCH ALGORITHM
# ============================================================

def dfs(problem, state, index):

    if problem.is_goal(index):

        print("\nSUCCESS: LOAD DISTRIBUTION COMPLETED")
        return True

    consumer = problem.consumers[index]

    print("\n------------------------------------------------")
    print("Processing Consumer:", consumer.consumer_id)
    print("Area:", consumer.area)
    print("------------------------------------------------")

    for node in problem.nodes:

        print("\nChecking Grid Node:", node)

        if problem.is_safe(node, consumer.area, state):

            print("Safe Allocation Possible")

            state.allocations[node].append(
                (consumer.consumer_id, consumer.area)
            )

            print(
                "Allocated",
                consumer.consumer_id,
                "to",
                node
            )

            # Recursive DFS Search

            if dfs(problem, state, index + 1):

                return True

            # =================================================
            # BACKTRACKING
            # =================================================

            removed_consumer = state.allocations[node].pop()

            print("Backtracking Performed")

            print(
                "Removed Consumer:",
                removed_consumer[0]
            )

        else:

            print("Constraint Violated")
            print("Cannot Allocate Consumer")

    return False

# ============================================================
# CO4 - UTILITY BASED DECISION MAKING
# ============================================================

def utility_selection(state):

    print("\n================================================")
    print("UTILITY BASED BEST NODE SELECTION")
    print("================================================")

    remaining_space = {}

    for node in state.allocations:

        remaining_space[node] = (
            GRID_NODES[node].capacity -
            len(state.allocations[node])
        )

    best_node = max(
        remaining_space,
        key=remaining_space.get
    )

    print("Best Grid Node:", best_node)
    print("Available Capacity:", remaining_space[best_node])

# ============================================================
# CO5 - BAYESIAN REASONING
# ============================================================

def bayesian_analysis():

    print("\n================================================")
    print("BAYESIAN LOAD FAILURE ANALYSIS")
    print("================================================")

    prob_failure = 0.3
    prob_overload = 0.7
    prob_detection = 0.8

    bayes_probability = (
        prob_detection * prob_failure
    ) / prob_overload

    print("Probability of Detecting Grid Failure:")
    print(round(bayes_probability, 2))

# ============================================================
# DISPLAY FUNCTIONS
# ============================================================

def show_consumers():

    print("\n================================================")
    print("CONSUMER DETAILS")
    print("================================================")

    for consumer in CONSUMERS:

        print(
            "Consumer ID:",
            consumer.consumer_id,
            "| Area:",
            consumer.area
        )

def show_nodes():

    print("\n================================================")
    print("GRID NODE DETAILS")
    print("================================================")

    for node_id, node in GRID_NODES.items():

        print(
            "Node:",
            node_id,
            "| Capacity:",
            node.capacity
        )

# ============================================================
# FINAL OUTPUT
# ============================================================

def show_final_distribution(state):

    print("\n================================================")
    print("FINAL LOAD DISTRIBUTION")
    print("================================================")

    for node in state.allocations:

        print("\nGrid Node:", node)

        if state.allocations[node]:

            for consumer in state.allocations[node]:

                print(
                    "Consumer:",
                    consumer[0],
                    "| Area:",
                    consumer[1]
                )

        else:

            print("No Consumers Allocated")

# ============================================================
# CO6 - INTEGRATED AI SYSTEM SUMMARY
# ============================================================

def show_summary():

    print("\n================================================")
    print("PROJECT SUMMARY")
    print("================================================")

    print("\nCO1 - AI Problem Representation")
    print("• Consumers and grid nodes represented using classes")
    print("• State representation implemented")

    print("\nCO2 - DFS Search")
    print("• DFS used for smart load allocation")
    print("• Recursive graph traversal implemented")

    print("\nCO3 - CSP + Backtracking")
    print("• Constraint checking implemented")
    print("• Backtracking performed during conflicts")

    print("\nCO4 - Utility Based Decision")
    print("• Best node selected using utility logic")

    print("\nCO5 - Bayesian Reasoning")
    print("• Probabilistic load failure analysis performed")

    print("\nCO6 - Integrated AI System")
    print("• Combined AI reasoning techniques")
    print("• Generated explainable smart grid solution")

# ============================================================
# MAIN PROGRAM
# ============================================================

def main():

    print("\n================================================")
    print("SMART GRID LOAD DECISION AGENT")
    print("AI BASED CFAI PROJECT")
    print("CO1 + CO2 + CO3 + CO4 + CO5 + CO6")
    print("================================================")

    # Display Input Data

    show_consumers()
    show_nodes()

    # Initial State

    initial_state = State(
        allocations={
            "G1": [],
            "G2": [],
            "G3": []
        }
    )

    # Create Problem

    problem = SmartGridProblem(
        CONSUMERS,
        GRID_NODES
    )

    print("\n================================================")
    print("SMART GRID AI PROCESS STARTED")
    print("================================================")

    # Start DFS Search

    success = dfs(
        problem,
        initial_state,
        0
    )

    # Final Output

    if success:

        print("\nLOAD DISTRIBUTION GENERATED SUCCESSFULLY")

        show_final_distribution(initial_state)

    else:

        print("\nFAILED TO GENERATE LOAD DISTRIBUTION")

    # CO4

    utility_selection(initial_state)

    # CO5

    bayesian_analysis()

    # CO6

    show_summary()

    print("\n================================================")
    print("PROJECT COMPLETED SUCCESSFULLY")
    print("================================================")

# ============================================================
# DRIVER CODE
# ============================================================

if __name__ == "__main__":

    main()
```
